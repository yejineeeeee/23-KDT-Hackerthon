```python
# import ultralytics
from ultralytics import YOLO
import torch
import os

model = YOLO("yolov8n.pt")  # load a pretrained model (recommended for training)

# Train the model
current_directory = os.getcwd()
file_path = os.path.join(current_directory, "tree.yaml")
model.train(data=file_path, epochs=100, patience=30, batch=64)

metrics = model.val()  # evaluate model performance on the validation set
```

```python
path = model.export(format="onnx")  # export the model to ONNX format
model_path = os.path.join(current_directory, "yolov8_tree.pt")
torch.save(model.state_dict(), model_path)
print(f"Model has been saved to {model_path}")
```
