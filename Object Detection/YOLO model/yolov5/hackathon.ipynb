```python
# 입력 JSON 디렉토리 경로
input_json_dir = "/content/Dataset/01.데이터/Training/02.라벨링데이터"

# 출력 TXT 디렉토리 경로
output_txt_dir = "/content/Dataset/01.데이터/Training/02.라벨링데이터"

# 이미지 해상도
image_width = 1280
image_height = 1280





# 입력 JSON 파일을 TXT 파일로 변환하여 저장
for json_file_name in os.listdir(input_json_dir):
    if json_file_name.endswith(".json"):
        # JSON 파일 읽기
        with open(os.path.join(input_json_dir, json_file_name), "r") as json_file:
            data = json.load(json_file)

        # 출력 TXT 파일 경로 생성
        txt_file_name = os.path.splitext(json_file_name)[0] + ".txt"
        txt_file_path = os.path.join(output_txt_dir, txt_file_name)
        yolo_labels = []

        annotations = data.get("annotations", {})
        bbox_list = annotations.get("bbox", [])
        with open(txt_file_path, "w") as txt_file:
            for bbox_info in bbox_list:
                label = bbox_info.get("label", "")
                if label in class_mapping:
                    label = class_mapping[label]
                    if label in range(29,46):
                        if annotations["class"] == "남자사람":
                            label+=18

                x_center_norm = (bbox_info.get("x") + bbox_info.get("w") / 2) / image_width
                y_center_norm = (bbox_info.get("y") + bbox_info.get("h") / 2) / image_height
                w_norm = bbox_info.get("w") / image_width
                h_norm = bbox_info.get("h") / image_height
                txt_file.write(f"{label} {x_center_norm:.6f} {y_center_norm:.6f} {w_norm:.6f} {h_norm:.6f}\n")
```

```python
import os
import shutil

# 원본 디렉토리 경로
source_directory = "/content/Dataset/01.데이터/Training/01.원천데이터"

# 대상 디렉토리 경로
target_directory = "/content/data/images/train"

# 원본 디렉토리 내의 모든 .jpg 파일 목록 가져오기
jpg_files = [file for file in os.listdir(source_directory) if file.endswith(".jpg")]

# 대상 디렉토리가 없으면 생성
os.makedirs(target_directory, exist_ok=True)

# .jpg 파일들을 대상 디렉토리로 이동
for jpg_file in jpg_files:
    source_path = os.path.join(source_directory, jpg_file)
    target_path = os.path.join(target_directory, jpg_file)
    shutil.move(source_path, target_path)

print("모든 .jpg 파일이 이동되었습니다.")
```
